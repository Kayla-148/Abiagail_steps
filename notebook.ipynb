### Your Highest Step Counts

```python
Max = steps.sort_values(by=['Stairs'], ascending=False).head(5)
Max.set_index('Date', inplace=True)
Max
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define fixed colors for different ranges of steps
def get_color(step):
    if step < 1500:
        return 'red'
    elif step < 2000 and step > 1500:
        return 'darkorange'
    elif step < 2500 and step > 2000:
        return 'yellow'
    else:
        return 'lime'

colors = steps['Stairs'].apply(get_color)

fig1, ax = plt.subplots(figsize=(15, 10))
bars = ax.bar(steps['Date'], steps['Stairs'], color=colors, edgecolor='black', width= 0.7)

plt.xticks(rotation=90)

font1 = {'family': 'serif', 'color': 'black', 'size': 25, 'weight': 'semibold'}
font2 = {'family': 'serif', 'color': 'black', 'size': 18}

ax.set_xlabel('Date', fontdict=font2)
ax.set_ylabel('Steps', fontdict=font2)
ax.set_title('Steps on a given date', fontdict=font1)

plt.grid(axis = 'y')

ax.annotate("", xy=[7.5,0], 
             xytext=(7.5,-200), arrowprops={'arrowstyle':'simple', 'color':'fuchsia'})

ax.annotate("", xy=[25.5,0], 
             xytext=(25.5,-200), arrowprops={'arrowstyle':'simple', 'color':'fuchsia'})

ax.annotate("", xy=[46.5,0], 
             xytext=(46.5,-200), arrowprops={'arrowstyle':'simple', 'color':'fuchsia'})

plt.show()
```

On this graph, dates where your steps were over 2,500 are green, between 2,000 and 2,500 are yellow, between 2,000 and 1,500 are orange, and less than 1500 are red. On all dates that are red you used the stairs for 20 minutes or less. On all other dates you used the Stairs for 35 minutes. The pink arrows between dates on the botton indicate gaps of more than 1 month in data. In some of these gaps you were still logging gym excersizes but they were for treadmill or elliptical.
  
During one of those time gaps at the beggining of the graph there is one month missing where I beleive we would have seen your improvment gradually move from orange to yellow.     

Other fluctuations in step count may be due to the other excersizes you were doing that day. For example, if you did 20 minutes on the elliptical and then 35 minutes on stairs you may do less steps because you're already tired from the previous workout.
